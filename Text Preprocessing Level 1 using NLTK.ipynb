```python
#importing required libraries
import nltk
import io
import random
import string
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
#reading the document
doc=open('E:\\donald.txt')
file=doc.read()
file
```

```python
#tokenizing file into sentences
from nltk import sent_tokenize
sentences=nltk.sent_tokenize(file)
sentences
```

```python
#tokenizing file into words
from nltk import word_tokenize
words=nltk.word_tokenize(file)
words
len(words)
```

```python
words
```

```python
#we have to remove punctuations from the words in order to clean the file
word=[word.lower() for word in words if word.isalpha()]
```

```python
#checking the word list
word
```

```python
#we have to remove stopwords from the words list 
from nltk.corpus import stopwords
cleaned_words=[]
for w in word:
    if w not in stopwords.words('english'):
        cleaned_words.append(w)
```

```python
#checking cleaned_words list
cleaned_words
```

```python
#we will find the top 5 repeated words
from nltk.probability import FreqDist
frq=FreqDist(cleaned_words)
frq.most_common(5)
```

```python
#plotting the top 5 words
frq.plot(5)
```

```python
#bulding wordcloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt
wc=WordCloud().generate(file)
plt.axis('off')
plt.imshow(wc)
```

```python
#converting each and evry word into root word using stemming technique
from nltk.stem import PorterStemmer
ps=PorterStemmer()
stemm=[ps.stem(word) for word in word]
    
    
```

```python
stemm
```

```python
#converting each and evry word into root word using lemmatization
from nltk.stem import WordNetLemmatizer
wl=WordNetLemmatizer()
wl=[wl.lemmatize(word) for word in word]
```

```python
wl
```

```python
#pos tagging
from nltk import pos_tag
parts_of_speech_lemm=pos_tag(wl)
```

```python
parts_of_speech_lemm
```

```python
parts_of_speech_stemm=pos_tag(stemm)
parts_of_speech_stemm
```

```python
#noun phrase chunking where we search for chunks corresponding to individual noun phrases.
#NP chunk should be formed whenever the chunker finds an optional determiner (DT) followed by any number of adjectives(JJ) and then a noun (NN)
pattern="NP: {<DT>?<JJ>*<NN>}"
chunkparser=nltk.RegexpParser(pattern)
result=chunkparser.parse(parts_of_speech_lemm)
print(result)
```

```python
result.draw()
```
